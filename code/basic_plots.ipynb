# Colour map

```python
tmax = 500
L = 500
N = 100
vmax = 5
p = 0.2

simulation = functions.Nagel_Schreckenberg(L, N, vmax, p, tmax)
simulation.run_model()

points = np.zeros((tmax, L))
for i in range(L):
    for j in range(tmax):
        if simulation.positions[j][i] > -1:
            points[j][i] = 1
        else:
            points[j][i] = 0

plt.figure(dpi=150)
plt.imshow(points, cmap = "cool", interpolation = "nearest")
plt.show()
```
